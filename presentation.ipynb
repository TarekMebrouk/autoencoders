# Réduction de dimensionnalité

## 1- Question
Plus on a de données stockées dans les bases, plus on a d’avantages dans l’exécution des algorithmes d’apprentissage automatique. Alors quel est donc l’intérêt de supprimer des variables prédictives ?

## 2- Définition
La réduction de dimensionnalité fait référence aux techniques qui réduisent le nombre de variables prédictives dans un ensemble de données, ou encore projettent des données issues d’un espace de grande dimension dans un espace de plus petite dimension. Ceci peut être effectué après le nettoyage et la normalisation des données et avant l’entraînement afin de détecter les colonnes capturant la partie essentielle de nos données.


**Remarque : Le nombre de variables prédictives (features) pour un set de données est appelé sa dimension.**

## 3-	Les avantages de la réduction de dimensionnalité



1.   ***Espace de stockage :*** La diminution du nombre de dimension implique la réduction de l’espace requis pour le stockage des données.

2.   ***Facteur de temps :*** Moins de dimensions entraînent moins de temps de calcul/apprentissage

3.   ***Performances des algorithmes d’apprentissage :*** 

*	Il améliore la précision du modèle en supprimant le bruit dans les données d’apprentissage avant l’entrainement.

*	Il supprime les fonctionnalités redondantes (problème de multi-colinéarité). Par exemple, on a deux variables : « le temps passé sur le tapis roulant » et « les calories brûlées ». Ces variables sont fortement corrélées car plus vous passez de temps à courir sur un tapis roulant, plus vous brûlerez de calories. Par conséquent, il est inutile de stocker les deux car un seul d'entre eux fait ce dont on a besoin.

*	Certains algorithmes ne fonctionnent pas bien lorsque nous avons une grande dimension. Il faut donc réduire ces dimensions pour que l'algorithme soit utile.

4.   ***Visualisation des données :*** Il est très difficile de visualiser des données dans des dimensions plus élevées, donc réduire l’espace en 2D ou 3D peut nous permettre de tracer et d'observer des modèles plus clairement.

5.   ***Séparation linéaire :*** La réduction de dimensionnalité peut être utilisée pour transformer des données non linéaires en une forme séparable linéairement.






## 4-	Techniques de réduction de dimensionnalité

1.	***Sélection de caractéristiques :*** 

Conserve que les fonctionnalités les plus importantes dans l'ensemble de données et supprime les fonctionnalités redondantes.
Exemple : Imaginons que notre ensemble de données comporte également une colonne sur les films préférés des clients, et nous nous rendons compte que cela ne nous aide pas énormément à prévoir combien ils dépenseront dans notre magasin. Avec la sélection de fonctionnalités, nous exclurions simplement toute cette fonctionnalité.


2.	***Extraction des caractéristiques :*** 

Conserve des parties de fonctionnalités, de sorte que nous ne conservons que les éléments les plus utiles de chaque fonctionnalité. De cette façon, nous ne perdons pas trop d'informations en supprimant des fonctionnalités entières. Le nouvel ensemble de fonctionnalités contient des valeurs différentes au lieu des valeurs d'origine.

## 6-	Encodeur décodeur (autoencoders)


***Définition :*** Est un réseau de neurones artificiels utilisé pour l'apprentissage non supervisé, qui possèdent exactement le même nombre de neurones sur leur couche d’entrée et leur couche de sortie. Le but pour un auto encodeur est d’avoir une sortie la plus proche de l’entrée (minimiser le coût de reconstruction entre la sortie et l’entrée).image_1.png


***Architecture et principe :***

1.	**Encoder :** 

Un module qui aide à « compresser » les données d'entrée dans une représentation codée qui est généralement inférieure de plusieurs ordres de grandeur aux données d'entrée.


2.	**Bottleneck :** 

Un module qui contient les représentations de connaissances compressées et qui est donc la partie la plus importante du réseau


3.	**Decoder :** 

Un module qui aide le réseau à « décompresser » les représentations des connaissances et reconstruit les données à partir de leur forme codée.


***Hyper paramètres :***


1.	**Taille du bottleneck :**  

Représente l'hyper paramètre le plus important utilisé pour régler l'auto-encodeur. Il détermine la quantité de données à compresser. 


2.	**Nombre de couches :** 

La profondeur de l'encodeur et du décodeur (une profondeur plus élevée augmente la complexité du modèle, une profondeur plus faible est plus rapide à traiter).


3.	**Nombre de nœuds par couche :** 

Représente les poids que nous utilisons par couche. En règle générale, le nombre de nœuds diminue avec chaque couche suivante dans l'auto-encodeur à mesure que l'entrée de chacune de ces couches devient plus petite à travers les couches. 


4.	**Fonction de perte :** 

La fonction dépend fortement du type d'entrée et de sortie auquel nous voulons que l'auto-encodeur s'adapte. (MSE mean of squared errors si on travaille sur des images par exemple).


***Autoencoder débruiteur :***


Après avoir appris comment représenter de manière optimale les entrées, en fournissant à l’auto encodeur des entrées bruitées, il est capable de donner en sortie des données débruitées qui sont très proches de l’entrée originale
